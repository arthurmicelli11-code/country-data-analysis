# Country Data Analysis

## Objective
Analyze country-level data to explore population, GDP, and regional economic patterns.


```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/country_data.csv")
df.head()
df.info()
df.describe()
df.isna().sum()
```

```python
# Top 10 countries by GDP
top_gdp = df.sort_values(by="gdp", ascending=False).head(10)
top_gdp[["country", "gdp"]]
```

```python
# Top 10 countries by population
top_population = df.sort_values(by="population", ascending=False).head(10)
top_population[["country", "population"]]
```

```python
df["gdp_per_capita"] = df["gdp"] / df["population"]
df.head()
```

```python
# Top GDP per capita
df.sort_values(by="gdp_per_capita", ascending=False).head(10)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure()
plt.scatter(df["population"], df["gdp"])
plt.xlabel("Population")
plt.ylabel("GDP")
plt.title("GDP vs Population")
plt.savefig("../visuals/gdp_vs_population.png")
plt.show()
```

```python
plt.figure()
plt.barh(top_gdp["country"], top_gdp["gdp"])
plt.xlabel("GDP")
plt.title("Top 10 Countries by GDP")
plt.savefig("../visuals/top_10_countries_by_gdp.png")
plt.show()
```

## Key Insights

- GDP per capita provides a better comparison than total GDP.
- Population size alone does not determine economic strength.
- Some countries with smaller populations outperform larger economies.
