# Country Data Analysis

## Objective
Analyze country-level data to explore population, GDP, and regional economic patterns.
This analysis focus to support strategic decision-making and comparative economic insights.

## Tools
- Python
- Pandas
- Matplotlib
- Jupyter Notebook


```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/country_data.csv")
df.head()
```

## Data Overview

Initial inspection of the dataset to understand structure, columns, and basic statistics.


```python
df.info()
df.describe()
```

## GDP Analysis

Analysis of total GDP and GDP per capita to compare economic strength between countries.


```python
top_gdp = df.sort_values(by="gdp", ascending=False).head(10)
top_gdp
```

```python
df["gdp_per_capita"] = df["gdp"] / df["population"]
df.sort_values(by="gdp_per_capita", ascending=False).head(10)
```

## Large Population Countries

Analysis focused on countries with population greater than 10 million.


```python
large_countries = df[df["population"] > 10_000_000]
large_countries.sort_values(by="gdp_per_capita", ascending=False).head(10)
```

## Visualization

Scatter plot showing the relationship between population and GDP.


```python
plt.figure(figsize=(8,5))
plt.scatter(df["population"], df["gdp"])
plt.xlabel("Population")
plt.ylabel("GDP")
plt.title("GDP vs Population by Country")
plt.show()
```

## Key Insights

- GDP per capita provides a more accurate comparison between countries than total GDP.
- Population size alone does not determine economic strength.
- Some smaller countries outperform larger ones economically.
